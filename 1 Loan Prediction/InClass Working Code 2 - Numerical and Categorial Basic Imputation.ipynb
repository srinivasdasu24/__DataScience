### We Combine both Datasets to perform Missing Value Imputation

```python
#CodeSection6
all_data = [ train, test]
```

```python
#CodeSection7
# Combine both using concat (with axis = 0) to check Missing Value
pd.concat([train, test] , axis=0).isnull().sum()
```

## 3/ Impute Numerical Variables
- Impute all Numerical Variables 

### Variable/Feature 1

```python
#CodeSection8
#ApplicantIncome    367 non-null    int64  
# 7   CoapplicantIncome  367 non-null    int64  
# 8   LoanAmount         362 non-null    float64
# 9   Loan_Amount_Term   361 non-null    float64
# 10  Credit_History     338 non-null    float64
# Impute Feature with mean/median for both train and test - use for loop
for data in all_data:
  data['ApplicantIncome'] = data['ApplicantIncome'].fillna(train['ApplicantIncome'].median())
```

### Variable/Feature 2

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for data in all_data:
  data['LoanAmount'] = data['LoanAmount'].fillna(train['LoanAmount'].median())
```

### Variable/Feature 3

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for data in all_data:
  data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].median())
```

### Variable/Feature 4

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for data in all_data:
  data['Credit_History'] = data['Credit_History'].fillna(train['Credit_History'].median())
```

## 4/ Impute Categorical Variables
- Impute all Numerical Variables

### Categorical Variable/Feature 1

```python
#CodeSection9
# Gender                24
# Married                3
# Dependents            25
# Self_Employed         55
# Impute Feature with mode (use [0]) for both train and test - use for loop
for data in all_data:
  data['Gender'] = data['Gender'].fillna(train['Gender'].mode()[0])
```

### Categorical Variable/Feature 2

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for data in all_data:
  data['Married'] = data['Married'].fillna(train['Married'].mode()[0])
```

### Categorical Variable/Feature 3

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for data in all_data:
  data['Dependents'] = data['Dependents'].fillna(train['Dependents'].mode()[0])
```

### Categorical Variable/Feature 4

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for data in all_data:
  data['Self_Employed'] = data['Self_Employed'].fillna(train['Self_Employed'].mode()[0])
```

## 5/ Now Re Check all of them

```python
#CodeSection10
# Combine both to check Missing Value
pd.concat([train, test], axis = 0).isnull().sum()
```

## 6/ Prepare Data for Model Model

```python
#CodeSection11
# Check first 5 rows of Test Dataset
test.head()
```

```python
#CodeSection12
# Divide Data into X and y
X = train.drop(['Loan_ID','Loan_Status'], axis =1)
Y = train['Loan_Status']
```

### Create Dummy Variables for  Categorical Variables

```python
#CodeSection13
# Create Dummy Variables
X = pd.get_dummies(X)
```

```python
#CodeSection14
# Check first 5 rows of X
X.head()
```

## 7/ Build Model

```python
#CodeSection15
#Start by importing LogisticRegression 
from sklearn.linear_model import LogisticRegression

#Initialize LogisticRegression
clf = LogisticRegression(max_iter=500)
```

```python
#CodeSection16
# Fit Model
model = clf.fit(X,Y)
```

### Check Accuracy of Model on Train Data

```python
#CodeSection17
# Predict on Train Data and check predictions
y_pred = model.predict(X)
y_pred
```

### We can use a Accuracy Function from Metrics

```python
#CodeSection17
# Import accuracy_score 
from sklearn.metrics import accuracy_score
```

```python
#CodeSection18
# Get Actual Predictions
y_true = train['Loan_Status']
```

```python
#CodeSection19
# Compare True values versus prediction we made
print(f" Train Accuracy : {accuracy_score(y_true, y_pred):0.1%}")
```
